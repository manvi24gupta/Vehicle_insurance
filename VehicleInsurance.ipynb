-------Statstical analysis using various regression models because we have to find various attributes
that depends on the value of clv(customer lifetime value) and then predict the further values ------
--- we will be conducting three regression models
1) MultiLinear regression : to check the linear dependency of clv on monthly premium auto , coverage ,number of policies , total claim amount and then applied gradient boosting 
2) Polynomial Regression : to the polynomial dependency of clv on monthly premium auto , coverage ,number of policies , total claim amount and then applied gradient boosting
3) Decision tree analysis to classify instances by sorting them down the tree to check the response decision.
4) PCA to reduce the number of components.

----MultiLinear Regression Model---

```python
from sklearn.linear_model import LinearRegression 
```

```python
X = df[['Coverage','Monthly.Premium.Auto','Number.of.Policies','Total.Claim.Amount']]
y = df['clv']
```

```python
model=LinearRegression().fit(X,y)
model
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2)
model.fit(X_train,y_train)
pred = model.predict(X_test)
```

```python
pred=pred.reshape(-1,1)
pred
```

```python
model.score(X_train,y_train)
```

```python
coef = pd.DataFrame(model.coef_,X.columns,columns=['coeff'])
coef 
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(y.index[:100],y[:100], label='Actual', color='b')
plt.plot(pred[:100], label='Predicted', color='r')
```

```python
from sklearn import ensemble
reg = ensemble.GradientBoostingRegressor( n_estimators = 100 ,learning_rate = 0.012)
reg.fit(X_train, y_train)
```

```python
reg.fit(X,y)
```

```python
pred = reg.predict(X)
pred = pd.DataFrame(pred)
pred.head()
```

```python
reg.score(X,y)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(y.index[:100], y[:100], label='Actual', color='b')
plt.plot(pred[:100], label='Predicted', color='r')
```

----PCA-----

```python
from sklearn.preprocessing import StandardScaler 
  
scalar = StandardScaler() 
  
scalar.fit(X) 
scaled_data = scalar.transform(X) 
```

```python
from sklearn.decomposition import PCA 
pca = PCA() 
pca.fit(scaled_data) 
```

```python
variance = pca.explained_variance_ratio_ 
variance
```

```python
var=np.cumsum(np.round(pca.explained_variance_ratio_, decimals=3)*100)
var
```

```python
pca = PCA(n_components=5) 
X_transformed = pca.fit_transform(X) 
```
